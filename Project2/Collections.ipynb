# Import required libraries

```python
df = pd.read_csv('charcters_stats.csv', index_col=0)
df.head()
```

# Find the amount of characters who fall under each alignment, you can use value_counts()

```python
df.value_counts("Alignment")
```

# Find only the good alignment characters

```python
df.loc[df["Alignment"] == "good"]
```

# Show only the first five records from the previous bullet point

```python
df.loc[df["Alignment"] == "good"].head(5)
```

# Select the top five heroes with the highest speed who have a good alignment

```python
df.loc[df["Alignment"] == "good"].sort_values(by = 'Speed', ascending=False).head(5)
```

# Select five heroes with maximum power who have a good alignment

```python
df.loc[df["Alignment"] == "good"].sort_values(by = 'Power', ascending=False).head(5)
```

# Find out how many superheroes there are with power of 100 with good alignment

```python
df.loc[(df["Alignment"] == "good") & (df["Power"] == 100)].sort_values(by = 'Power', ascending=False).count()
```

# Draw a bar plot of all superheroes who are having good alignment and a max power of the top five heroes only , take the same object of point 13, show name and total in plot with green bars
