#	Load the csv file and show top 5 records from it

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('players_20.csv', index_col=0)

df.head(5)
```

#	How you would be able to see each column's name

```python
df.keys()
```

#	Show the number of rows and columns in the dataset

```python
len(df)
```

```python
len(df.columns)
```

#	Show the number of players in each country

```python
df.groupby('nationality').count()
```

#	Show the top 10 countries with the highest amount of players

```python
df.groupby('nationality').count().sort_values(by = 'player_url', ascending = False).head(10)
```

#	Create a bar plot of the top 5 countries with the highest number of players, and fill the bars in green

```python
df.groupby('nationality').count().sort_values(by = 'player_url', ascending = False).head(5)['player_url'].plot.bar(color='green')
```

#	Display the short name and wages for the top 5 players

```python
df.sort_values(by = 'value_eur', ascending = False).head(5)[['short_name', 'wage_eur']]
```

#	Display the short name and wages for the top 5 players with the highest salaries

```python
df.sort_values(by = 'wage_eur', ascending = False).head(5)[['short_name', 'wage_eur']]
```

#	Create a bar plot of the previous bullet point

```python
df.sort_values(by = 'wage_eur', ascending = False).head(5)[['short_name', 'wage_eur']].plot.bar(x='short_name', y = 'wage_eur')
```

#	Show top 10 records for Germany

```python
df.groupby('nationality').get_group('Germany').sort_values(by = 'weight_kg', ascending=False).head(5)
```

```python
df.groupby('nationality').get_group('Germany').sort_values(by = 'wage_eur', ascending=False).head(5)
```
